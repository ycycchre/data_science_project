```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
```

```python
lysis_data=pd.read_csv("./Stress-Lysis.csv")
print(lysis_data.head())
```

```python
# print(lysis_data[['Humidity', 'Temperature']])
temp_data = lysis_data['Temperature']
humi_data = lysis_data['Humidity']
step_data = lysis_data['Step count']
str_lev_data = lysis_data['Stress Level']
```

```python
humi_stress_correlation = humi_data.corr(str_lev_data)
print("correlation coefficient between humidity and stress level is", humi_stress_correlation)

plt.scatter(lysis_data['Humidity'], lysis_data['Stress Level'], alpha=0.8)
plt.xlabel('Humidity')
plt.ylabel('Stress Level')
plt.title("Correlation = {}".format(humi_stress_correlation))
plt.show()
```

```python
temp_stress_correlation = temp_data.corr(str_lev_data)
print("correlation coefficient between temperature and stress level is", temp_stress_correlation)

plt.scatter(lysis_data['Temperature'], lysis_data['Stress Level'], alpha=0.8)
plt.xlabel('Temperature')
plt.ylabel('Stress Level')
plt.title("Correlation = {}".format(temp_stress_correlation))
plt.show()
```

```python
step_stress_correlation = step_data.corr(str_lev_data)
print("correlation coefficient between step count and stress level is", step_stress_correlation)

plt.scatter(lysis_data['Step count'], lysis_data['Stress Level'], alpha=0.8)
plt.xlabel('Step count')
plt.ylabel('Stress Level')
plt.title("Correlation = {}".format(step_stress_correlation))
plt.show()
```

```python
temp_step_correlation = temp_data.corr(step_data)
print("correlation coefficient between temperature and step count is", temp_step_correlation)

plt.scatter(lysis_data['Temperature'], lysis_data['Step count'], alpha=0.8)
plt.xlabel('Temperature')
plt.ylabel('Step count')
plt.title("Correlation = {}".format(temp_step_correlation))
plt.show()
```

```python
temp_humi_correlation = temp_data.corr(humi_data)
print("correlation coefficient between temperature and humidity is", temp_humi_correlation)

plt.scatter(lysis_data['Temperature'], lysis_data['Humidity'], alpha=0.8)
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.title("Correlation = {}".format(temp_humi_correlation))
plt.show()
```

```python
humi_step_correlation = humi_data.corr(step_data)
print("correlation coefficient between humidity and step count is", humi_step_correlation)

plt.scatter(lysis_data['Humidity'], lysis_data['Step count'], alpha=0.8)
plt.xlabel('Humidity')
plt.ylabel('Step count')
plt.title("Correlation = {}".format(humi_step_correlation))
plt.show()
```
