```python
import pandas as pd

import matplotlib.pyplot as plt
import numpy as np
```

```python
lysis_data=pd.read_csv("./Stress-Lysis.csv")
print(lysis_data.head())
```

```python
plt.scatter(lysis_data['Humidity'], lysis_data['Stress Level'], alpha=0.8)
plt.xlabel('Humidity')
plt.ylabel('Stress Level')
plt.show()
```

```python
plt.scatter(lysis_data['Temperature'], lysis_data['Stress Level'], alpha=0.8)
plt.xlabel('Temperature')
plt.ylabel('Stress Level')
plt.show()
```

```python
plt.scatter(lysis_data['Step count'], lysis_data['Stress Level'], alpha=0.8)
plt.xlabel('Step count')
plt.ylabel('Stress Level')
plt.show()
```

```python
stress_label = lysis_data['Stress Level']

x0 = lysis_data[stress_label == 0]
x1 = lysis_data[stress_label == 1]
x2 = lysis_data[stress_label == 2]
plt.scatter(x0.iloc[:, 0], x0.iloc[:, 2], c="red", marker='o', label='label0')
plt.scatter(x1.iloc[:, 0], x1.iloc[:, 2], c="blue", marker='*', label='label1')
plt.scatter(x2.iloc[:, 0], x2.iloc[:, 2], c="yellow", marker='+', label='label2')

plt.xlabel('Humidity')
plt.ylabel('Step count')

plt.show()
```
